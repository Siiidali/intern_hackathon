## Building the model

```python
import pandas as pd
import numpy as np
import torch
from torch_geometric.data import Data
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
import torch.optim as optim
```

```python
features = updated_df.iloc[:, 1:-3].values  # Assuming the last 3 columns are targets
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
node_features = torch.tensor(features_scaled, dtype=torch.float)
```

```python
# Encode class labels
targets = updated_df.iloc[:, -3:].idxmax(axis=1)
class_labels = pd.Categorical(targets).codes
class_labels_tensor = torch.tensor(class_labels, dtype=torch.long)
```

```python
len(targets)
```

```python
# Define the graph structure using nearest neighbors
nbrs = NearestNeighbors(n_neighbors=6, algorithm='auto').fit(features_scaled)  # Including self
distances, indices = nbrs.kneighbors(features_scaled)

edges = []
for i in range(indices.shape[0]):
    for j in range(1, indices.shape[1]):  # Exclude self
        edges.append([i, indices[i, j]])
edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
```

```python
# Split the data
num_nodes = len(updated_df)
indices = torch.randperm(num_nodes)
train_size = int(0.8 * num_nodes)

train_mask = torch.zeros(num_nodes, dtype=torch.bool)
test_mask = torch.zeros(num_nodes, dtype=torch.bool)

train_mask[indices[:train_size]] = True
test_mask[indices[train_size:]] = True

data.train_mask = train_mask
data.test_mask = test_mask
```

```python
class GCN(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_features, 16)
        self.conv2 = GCNConv(16, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
# Initialize the model and optimizer
model = GCN(num_features=node_features.shape[1], num_classes=3)  # Assuming 3 classes
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training function
def train():
    model.train()
    optimizer.zero_grad()
    out = model(data)
    loss = F.nll_loss(out[data.train_mask], class_labels_tensor[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss.item()

# Evaluation function
def evaluate():
    model.eval()
    out = model(data)
    _, pred = out[data.test_mask].max(dim=1)
    accuracy = accuracy_score(class_labels_tensor[data.test_mask].cpu(), pred.cpu())
    precision, recall, f1, _ = precision_recall_fscore_support(class_labels_tensor[data.test_mask].cpu(), pred.cpu(), average='weighted')
    return accuracy, precision, recall, f1
```

```python
node_features.shape[1]
```

```python
# Train and evaluate the model
for epoch in range(200):  # Adjust epochs as needed
    loss = train()
    if epoch % 20 == 0:  # Evaluation frequency
        accuracy, precision, recall, f1 = evaluate()
        print(f'Epoch {epoch}: Loss: {loss:.4f}, Acc: {accuracy:.4f}, Prec: {precision:.4f}, Rec: {recall:.4f}, F1: {f1:.4f}')
```
